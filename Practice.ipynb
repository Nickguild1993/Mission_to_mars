```python
# import Splinter and Beautiful Soup

from splinter import Browser
from bs4 import BeautifulSoup as soup
```

What the code below is doing:

We're creating an instance of a Splinter browser. Meaning, we're prepping our automated browser. 
We're also specifying that we'll be using * Chrome * as our browser.

* executable_path * is unpacking the dictionary we've stored the path in- like unpacking a suitcase.

* headless = False * means that all of the browser's actions wil be displayed in a CHROME window so we can see them.

```python
# because i'm using a MAC

executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
#### 10.3.2 #### PRACTICE WITH SPLINTER AND BEAUTIFULSOUP ####
```

```python
# Scape the Title 

# giving SPLINTER the URL we want to visit, which it'll do in chrome b/c we specified. .visit(url) is to 
# visit function, calling the (url)

# Visit the Quotes to Scrape site
url = "http://quotes.toscrape.com/"
browser.visit(url)
```

```python
# Use beautifulsoup to parse the HTML.  

# Parse the HTML

html = browser.html
html_soup = soup(html, "html.parser")

# this parses all the HTML on the page (ref: url). Meaning that BeautifulSoup has taken a look at the different
# components and can now access them.  

# Specifically, BeautifulSoup parses the HTML text and then stores it as an object.

# in our code, we're using "HTML.PARSER" to parse the info, but there are options available.
```

```python
# Find "title" and extract it!

# Scrape the title

title = html_soup.find("h2").text
title

# What that code did:
# 1). we used our *html_soup* object we created and chained *FIND()* to it search for the <h2/> tag.

# 2). we also extract ONLY THE TEXT within the HTML tags by adding **.text** to the end of the code.

## Output = 'Top Ten tags'
```

```python
# Scape all the tags. of which, there are 10.
tag_box = html_soup.find("div", class_="tags-box")

#tag_box
tags = tag_box.find_all("a", class_="tag")

for tag in tags:
    word = tag.text
    print(word)

    
# WHAT THE ABOVE CODE IS DOING #

# The first line: * tag_box = html_soup.find("div", class_="tags-box") *  creates a NEW VARIABLE: tag_box
# This new variable (tag_box) will be used to store the results of the search.
#         In this case, we're looking for: <div /> ELEMENTS with a class of: tag-box 
#          And we're searching for it in the HTML we parsed earlier and stored in the VARIABLE: html_soup.

# the SECOND line: tags = tag_box.find_all("a", class_="tag") 

# is similar to the first, but with a few tweaks to make it more specific. The new VARIABLE "tags" 
# will hold the results of a * find_all * but this time we're searching through the (already parsed directly above)
# results stored in our VARIABLE * tag_box * to find <a /> ELEMENTS with a "tag" class.

# ** used * find_all * this time b/c we want to capture ALL results, instead of a single or specific one. **

# Next, the *FOR* Loop.

# This FOR LOOP ycles through each tag in the variable: TAGS, strips the HTML code out of it, and then prints
# only the TEXT of each tag.

# for tag in TAGS: -> a FOR LOOP that cycles through each tag in the list
# word = tag.text -> Strips the HTML from the code and assigns the result to a variable 
# print(word) -> Prints each word in the list

# Output = love, inspirational, life humor,books, reading, friendship, friends, truth, simile
```

```python
# SCRAPE ACROSS MULTIPLE PAGES # (still 10.3.2)

# Code below will SCRAPE the quotes on the first page, click the "NEXT" button, then scrape more quotes 
# and so on (5 pages worth of quotes in total)

# When it's best to use a FOR LOOP -> Best to use FOR LOOP when we want code to be reapeted for a CERTAIN NUMBER
# of iterations.  
##### So in this case, want to use a FOR LOOP to scrape a page of quoest, then scrape the next page, and so on.
```

```python
# Scape acrosss pages #
url = "http://quotes.toscrape.com/"
browser.visit(url)

# again, this does two things- 
# 1st, assigns an actual URL to the variable named "url" 
# 2nd, tells Splinter to visit that webpage.
```

```python
# Creating FOR LOOP to collect each quote, "click" the next button, then collect the next set of quotes. 
# We'll use range(1, 6) in our for loop to visit the first five pages of the website.

for x in range(1,6):
    quote_soup = soup (html, "html.parser")
    quotes = quote_soup.find_all("span", class_="text")
    for quote in quotes:
        print("page:", x, "---------")
        print(quote.text)
    browser.links.find_by_partial_text("Next")
```
